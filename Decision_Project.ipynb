```python
import pandas as pd
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
```

```python
train = pd.read_csv(r"titanic\train.csv")
train
```

```python
train = train.drop(['PassengerId','Name','SibSp','Parch','Ticket','Embarked','Cabin','Fare'],axis="columns")
```

```python
train
```

```python

le_se = LabelEncoder()
```

```python

train['sex'] = le_se.fit_transform(train['Sex'])
```

```python
train.head()
```

```python
train = train.drop(['Sex'],axis="columns")
```

```python
train.head()
```

```python

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(train[['Pclass','sex','Age']],train['Survived'],test_size=0.2)
```

```python
test = pd.read_csv(r'titanic\test.csv')
```

```python
test.head()
```

```python
test = test.drop(['PassengerId','Name','SibSp','Parch','Ticket','Embarked','Cabin','Fare'],axis="columns")
```

```python
test['sex'] = le_se.fit_transform(test["Sex"])
```

```python
test
```

```python
model = tree.DecisionTreeClassifier()
```

```python
train.dropna(inplace=True)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test) 
```
